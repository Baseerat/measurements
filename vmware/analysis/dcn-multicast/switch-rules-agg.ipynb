```r
require(ggplot2)
require(data.table)
require(Hmisc)
require(ggthemes)
require(gridExtra)
require(grid)
require(lattice)
require(scales)
require(RColorBrewer)
```

```r
directory = "../../results/dcn-multicast/"
blues = brewer.pal(n = 6, "Blues")[4:6]
plotTheme <- theme_hc() + theme(legend.title=element_blank(), legend.position="none", legend.text=element_text(size=8))
```

```r
switchRuleAgg <- fread(paste(directory, "switch-rules-agg.csv", sep="/"), sep=",", header=TRUE)
```

```r
p1 <- ggplot(switchRuleAgg, aes(y=Rules, x=factor(Placement), fill=factor(Scheme))) +
 stat_summary(fun.data=mean_sdl, fun.args = list(mult=1), geom="bar", position=position_dodge(0.6), 
              width=0.6, col="black") + 
 ylab("No. of Switch Rules") +
 xlab("Placement") +
 scale_color_tableau() +
 scale_fill_manual(values=blues, name="Scheme") +
 plotTheme + theme(text=element_text(size=14), legend.title=element_text(size=12), legend.key.size = unit(0.25, "cm"))
```
